#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
data_0 = pd.read_csv("Students_info_0.csv", sep=",")
data_1 = pd.read_csv("Students_info_1.csv", sep=",")
data_2 = pd.read_csv("Students_info_2.csv", sep=",")
data_3 = pd.read_csv("Students_info_3.csv", sep=",")
data_4 = pd.read_csv("Students_info_4.csv", sep=",")
data_5 = pd.read_csv("Students_info_5.csv", sep=",")
data_6 = pd.read_csv("Students_info_6.csv", sep=",")
data_7 = pd.read_csv("Students_info_7.csv", sep=",")
data_8 = pd.read_csv("Students_info_8.csv", sep=",")
data_9 = pd.read_csv("Students_info_9.csv", sep=",")

data_info = pd.concat([data_0, data_1, data_2, data_3, data_4,
                       data_5, data_6, data_7, data_8, data_9,])
```

```python
marks_0 = pd.read_csv("Students_marks_0.csv", sep=",")
marks_1 = pd.read_csv("Students_marks_1.csv", sep=",")
marks_2 = pd.read_csv("Students_marks_2.csv", sep=",")
marks_3 = pd.read_csv("Students_marks_3.csv", sep=",")
marks_4 = pd.read_csv("Students_marks_4.csv", sep=",")
marks_5 = pd.read_csv("Students_marks_5.csv", sep=",")
marks_6 = pd.read_csv("Students_marks_6.csv", sep=",")
marks_7 = pd.read_csv("Students_marks_7.csv", sep=",")
marks_8 = pd.read_csv("Students_marks_8.csv", sep=",")
marks_9 = pd.read_csv("Students_marks_9.csv", sep=",")

data_marks = pd.concat([marks_0, marks_1, marks_2, marks_3, marks_4, 
                       marks_5, marks_6, marks_7, marks_8, marks_9,])
```

```python
data = data_info.merge(data_marks)
data
```

#### Задание 15. Сколько студентов успешно сдали все экзамены?

Создайте столбец OverAll_PassStatus и запишите в него для каждого студента 'F', если студент не сдал хотя бы один из трех экзаменов, а иначе 'P'.

Посчитайте количество студентов, которые сдали все экзамены.

```python
%%time
data['OverAll_PassStatus'] = 'F'
for i in range(1000):
    if ((data['math score'][i] > 50) & (data['reading score'][i] > 50) & (data['writing score'][i] > 50)):
        data['OverAll_PassStatus'][i] = 'P'
        
sum(data['OverAll_PassStatus'] == 'P')
```

#### Задание 16. Переведем баллы в оценки

### Система перевода баллов в оценки
####    больше 90 = A
####      80-90 = B
####      70-80 = C
####      60-70 = D
####      50-60 = E
####    меньше 50 = F (Fail)

Создайте вспомогательную функцию, которая будет по среднему баллу за три экзамена выставлять оценку студенту по данным выше критериям.

Создайте столбец Grade и запишите в него оценку каждого студента.

Выведите количество студентов, получивших каждую из оценок.

**В случае, если средний балл попадает на границу между оценками (т.е. равен ровно 60, 70 или 80 баллов), вы можете интерпретировать условие на своё усмотрение (т.е. можете поставить за 60 баллов оценку D, а можете - E).**

```python
data['average_mark'] = data[['math score','reading score','writing score']].sum(axis=1) / 3
data
```

```python
def Grade(average_mark):
    if  average_mark > 90:
        return "A"
    if  average_mark > 80:
        return "B"
    if  average_mark > 70:
        return "C"
    if  average_mark > 60:
        return "D"
    if  average_mark > 50:
        return "E"
    return 'F'    



data['grade'] = data['average_mark'].apply(Grade)
data
```
